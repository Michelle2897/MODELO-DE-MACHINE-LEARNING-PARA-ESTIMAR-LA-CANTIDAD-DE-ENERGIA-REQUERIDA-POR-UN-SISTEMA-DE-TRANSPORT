```python
# Librerías necesarias para predicción y análisis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import load_model

# Cargar el modelo guardado
ruta_modelo = r'C:\Users\Equipo\Downloads\modelo_autoencoder_lstm.h5'
autoencoder = load_model(ruta_modelo)
autoencoder.summary()


# Cargar y procesar el dataset
df_catama1 = pd.read_csv("D:/MAESTRIA BIG DATA/TESIS/BASES/Base_final_1.7_baja.csv", encoding='latin1', sep=';')
df_catama1['Fecha'] = pd.to_datetime(df_catama1['Fecha'], format='%d/%m/%Y', errors='coerce').dt.date
df_catama1.columns = df_catama1.columns.str.strip().str.replace('\n', ' ').str.replace(' +', ' ', regex=True)
columnas_a_eliminar = ['Dia', 'Semana', 'Mes', 'Año', 'Hora', 'Clave']
df_catama1 = df_catama1.drop(columns=[col for col in columnas_a_eliminar if col in df_catama1.columns])


# Convertir columnas a tipos adecuados
columnas_a_convertir_float = [
    'RADIACION Irradiación normal (MWh/m^2)', 'Temperature C',
    'Relative Humidity %', 'Precipitation Water mm',
    'Wind Speed (Velocidad del viento) m/s',
    'Wind Direction (Direccion del Viento) (deg)', 'Fotovoltáica MWh'
]
df_catama1[columnas_a_convertir_float] = df_catama1[columnas_a_convertir_float].replace(',', '.', regex=True).astype('float64')
df_catama1['Termica MWh (unidad 9)'] = pd.to_numeric(df_catama1['Termica MWh (unidad 9)'].replace(',', '.', regex=True), errors='coerce').round(3)
df_catama1['Energia Mwh'] = pd.to_numeric(df_catama1['Energia Mwh'].replace(',', '.', regex=True), errors='coerce').round(5)
df_catama1['Turistas'] = pd.to_numeric(df_catama1['Turistas'].replace(',', '.', regex=True), errors='coerce').fillna(0).astype('int64')

# Agrupar por fecha
df_agrupado = df_catama1.groupby('Fecha').sum().reset_index()

# Escalado de datos
input_features = ['RADIACION Irradiación normal (MWh/m^2)', 'Temperature C',
                  'Relative Humidity %', 'Precipitation Water mm',
                  'Wind Speed (Velocidad del viento) m/s',
                  'Wind Direction (Direccion del Viento) (deg)',
                  'Termica MWh (unidad 9)', 'Fotovoltáica MWh', 'Turistas']
target = 'Energia Mwh'

X = df_agrupado[input_features]
y = df_agrupado[target]

scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1))

# Agregar señal estacional
def add_seasonal_signal(X, period=365):
    t = np.arange(X.shape[0])
    seasonal_signal = np.sin(2 * np.pi * t / period).reshape(-1, 1)
    return np.concatenate([X, seasonal_signal], axis=1)

X_scaled = add_seasonal_signal(X_scaled)

# Crear secuencias
def create_sequences(X, y, time_steps=30):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps - 1])
    return np.array(Xs), np.array(ys).reshape(-1, 1)

time_steps = 30
X_seq, y_seq = create_sequences(X_scaled, y_scaled, time_steps)

# Hacer predicciones
y_pred_scaled_full = autoencoder.predict(X_seq)
y_pred_full = scaler_y.inverse_transform(y_pred_scaled_full[:, -1].reshape(-1, 1))
y_true_full = scaler_y.inverse_transform(y_seq[:, -1].reshape(-1, 1))

# Cálculo de métricas
mae_full = mean_absolute_error(y_true_full, y_pred_full)
rmse_full = np.sqrt(mean_squared_error(y_true_full, y_pred_full))
r2_full = r2_score(y_true_full, y_pred_full)

print(f"Mean Absolute Error (Full): {mae_full}")
print(f"Root Mean Squared Error (Full): {rmse_full}")
print(f"R^2 Score (Full): {r2_full}")

# Graficar Energía Real vs Predicha
plt.figure(figsize=(12, 6))
plt.plot(df_agrupado['Fecha'][time_steps:], y_true_full, label='Energía Real (Full)', color='blue', linestyle='-', alpha=0.7)
plt.plot(df_agrupado['Fecha'][time_steps:], y_pred_full, label='Energía Predicha por Autoencoder-LSTM (Full)', color='red', linestyle='--', alpha=0.7)
plt.xlabel('Fecha')
plt.ylabel('Energía (MWh)')
plt.title('Predicciones del Autoencoder vs Valores Reales para Todo el Conjunto de Datos')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Preparar datos para la predicción futura
# Generar Características Futuras
def generate_future_features(df, scaler_X, period=365, steps_ahead=365):
    last_known_inputs = df[input_features].values[-period:]  # Último año de datos conocido (un año de datos)
    future_inputs = np.tile(last_known_inputs, (steps_ahead // period + 1, 1))[:steps_ahead] #Genera un conjunto de datos futuros replicando los últimos datos conocidos
    future_inputs_scaled = scaler_X.transform(future_inputs) #Escala estos datos futuros
    future_inputs_scaled = add_seasonal_signal(future_inputs_scaled, period=period) #Añade una señal estacional a las características escaladas para capturar patrones estacionales
    return future_inputs_scaled

future_inputs_scaled = generate_future_features(df_agrupado, scaler_X, period=365, steps_ahead=365) 


# Predecir Energía Futura
def predict_future_energy(autoencoder, X_init, steps_ahead=365):
    predictions = []
    current_input = X_init[-1] # Inicializa con el último dato de la secuencia
    for step in range(steps_ahead): # Itera para cada paso futuro
        current_input_expanded = np.expand_dims(current_input, axis=0) #Expande la dimensión del current_input para que sea compatible con la entrada del modelo.
        predicted_energy_scaled = autoencoder.predict(current_input_expanded) #Predice la energía futura usando el modelo autoencoder.
        predicted_energy = scaler_y.inverse_transform(predicted_energy_scaled[:, -1].reshape(-1, 1)) # Desescala la predicción para obtener el valor real
        predictions.append(predicted_energy[0, 0]) #Almacena la predicción
        next_input = np.roll(current_input, -1, axis=0) #Reemplazar el valor más antiguo con la nueva entrada futura
        next_input[-1, :] = future_inputs_scaled[step] #Actualiza el último valor de la secuencia con la característica futura escalada.
        next_input[-1, -1] = predicted_energy_scaled[0, -1].item() #Actualiza el valor de la energía predicha en la última posición de la secuencia
        current_input = next_input #Actualiza current_input para el siguiente paso de predicción.
    return np.array(predictions) #Devuelve un array con todas las predicciones futuras

predictions = predict_future_energy(autoencoder, X_seq, steps_ahead=365) # genera predicciones de energía para cada uno de los próximos 365 días basándose en el modelo entrenado y los datos actuales

future_dates = pd.date_range(df_agrupado['Fecha'].iloc[-1] + pd.Timedelta(days=1), periods=365, freq='D')

plt.figure(figsize=(12, 6))
plt.plot(df_agrupado['Fecha'][time_steps:], y_true_full, label='Energía Real (Full)', color='blue', linestyle='-', alpha=0.7)
plt.plot(future_dates, predictions, label='Predicciones de Energía (MWh)', color='red', linestyle='--', alpha=0.7)
plt.xlabel('Fecha')
plt.ylabel('Energía (MWh)')
plt.title('Predicciones de Energía para un Año Futuro')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Crear un DataFrame para las fechas futuras
last_date = df_agrupado['Fecha'].max()
future_dates = [last_date + pd.DateOffset(days=i) for i in range(1, 366)]  # 365 días futuros
future_dates_df = pd.DataFrame(future_dates, columns=['Fecha'])

# Crear un DataFrame para las proyecciones futuras
future_predictions_df = pd.DataFrame({
    'Fecha': future_dates,
    'Predicción Energía MWh': predictions
})

# Verificar los valores de las proyecciones futuras
print(future_predictions_df.head(50))  # Muestra los primeros valores para verificar
```
